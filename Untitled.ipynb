```python
import numpy as np
import matplotlib.pyplot as plt

import pywt
import pywt.data


# Load image
from PIL import Image
import numpy as np
import os
import random

def get_image_from(path):
    images = os.listdir(path)
    total_images = len(images)
    index = random.randint(0, total_images)
    return os.path.join(path, images[index])
    
    
# get some random image
path = 'fake'


# load image
original = Image.open(get_image_from('./data_ver1/real')).convert('L')
original = np.array(original)

# Wavelet transform of image, and plot approximation and details
titles = ['Approximation', ' Horizontal detail',
          'Vertical detail', 'Diagonal detail']
coeffs2 = pywt.dwt2(original, 'db37')
LL, (LH, HL, HH) = coeffs2
print(LL.shape)
fig = plt.figure(figsize=(12, 3))
for i, a in enumerate([LL, LH, HL, HH]):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(a, interpolation="nearest", cmap=plt.cm.gray)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.tight_layout()

plt.show()
```
